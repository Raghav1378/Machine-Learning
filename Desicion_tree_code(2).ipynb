```python
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor 
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('BostonHousing (1).csv')
```

```python
X=df.iloc[:,:-1]
```

```python
Y=df.iloc[:,-1]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
DTR=DecisionTreeRegressor(max_depth=4,min_samples_leaf=2)
```

```python
DTR.fit(X_train,Y_train)
```

```python
DTR2=DecisionTreeRegressor(max_depth=3,min_samples_leaf=4)
```

```python
Y_pred_train=DTR.predict(X_train) 
Y_pred_test=DTR.predict(X_test)
```

```python
r2_score(Y_pred_train,Y_train)
```

```python
r2_score(Y_pred_test,Y_test)
```

```python
DTR2.fit(X_train,Y_train)
```

```python
Y_Pred_Train2=DTR2.predict(X_train)
Y_Pred_test2=DTR2.predict(X_test)
```

```python
r2_score(Y_train,Y_Pred_Train2)
```

```python
r2_score(Y_test,Y_Pred_test2)
```

```python
 DTR3=DecisionTreeRegressor(criterion='squared_error',max_depth=5)
```

```python
DTR3.fit(X_train,Y_train)
```

```python
p_test=DTR3.predict(X_test)
```

```python
r2_score(Y_test,p_test)
```
