```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score 
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC 
import pandas as pd
```

```python
X,Y=make_classification(n_samples=10000,n_features=10,n_informative=3)
```

```python

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
dt=DecisionTreeClassifier(random_state=42)
dt.fit(X_train,Y_train)
Y_pred=dt.predict(X_test)
```

```python
print("The accuracy score of the DTC is ",accuracy_score(Y_test,Y_pred))
```

# Bagging 

```python
bag=BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=500,max_samples=0.25,bootstrap=True,random_state=42,n_jobs=-1)
```

```python
bag.fit(X_train,Y_train)
```

```python
Y_pred2=bag.predict(X_test)
```

```python
print("The accuracy score of the BAgging is:",accuracy_score(Y_test,Y_pred2))
```

# BAGGING IN SVM

```python
bag2=BaggingClassifier(estimator=SVC(),bootstrap=True,n_estimators=500,max_samples=0.25)
```

```python
bag2.fit(X_train,Y_train)
```

```python
Y_pred3=bag2.predict(X_test)
```

```python
print("THe accuracy score of the bagging in SVM is: ",accuracy_score(Y_test,Y_pred3))
```

# OOB SCORE

# lets suppose you are sampling with replacement there might be chances that some of the data cannot come into the training anytime so that mean out of the bag sample

```python
bag=BaggingClassifier(
    estimator=DecisionTreeClassifier(),bootstrap=True,n_estimators=500,max_samples=0.25,oob_score=True,random_state=42,n_jobs=-1)
```

```python
bag.fit(X_train,Y_train)
```

```python
bag.oob_score_
```

```python
Y_pred4=bag.predict(X_test)
print("the accuracy score is ",accuracy_score(Y_pred4,Y_test))
```
